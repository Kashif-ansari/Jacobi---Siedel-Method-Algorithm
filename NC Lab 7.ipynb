# Jacobi Method

```python
import math as m
x = 0
y = 0
z = 0


for i in range(25):
    xt = (5-y+z)/4
    xt = round(xt,4)
    yt = (-4+x-z)/3
    yt = round(yt,4)
    zt = (1-(2*x)-(2*y))/5
    zt = round(zt,4)
    print(i+1)
    print("X = ",xt)
    print("Y = ",yt)
    print("Z = ",zt)
    print()
    if abs(x - xt)<0.0001:
        break
    else:
        x = xt
        y = yt
        z = zt
        
```
